```python
import torch 
from torch import nn
import matplotlib.pyplot as plt 
```

```python
weight = 0.7
bias = 0.3

X = torch.arange(0,1,0.02).unsqueeze(dim=1)
Y = X*weight + bias

```

```python
train_split = int(len(X)*0.8)

X_train, Y_train = X[:train_split], Y[:train_split]
X_test, Y_test = X[train_split:], Y[train_split:]

```

#### Creating the class of the nn Model 

```python
class linearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.weight = nn.Parameter(torch.randn(1, dtype=torch.float, requires_grad=True))
        self.bias = nn.Parameter(torch.randn(1, dtype=torch.float, requires_grad=True))

    def forward(self, x):
        return self.weight * x + self.bias

    
```

#### Creating the instance of it and making prediciton

```python
torch.manual_seed(42)
model2 = linearRegressionModel()

list(model2.parameters())


```

```python
with torch.inference_mode():
    y_preds = model2(X_test)

len(y_preds)


plt.plot(X,Y)
plt.plot(X_test, y_preds)
plt.show()
```

```python
lossFn = torch.nn.L1Loss()
optimizer = torch.optim.SGD(model2.parameters(), lr=0.001)


epochs = 800 

for epoch in range(epochs):
    model2.train()

    Y_preds = model2(X_train)

    loss_calculated = lossFn(Y_preds, Y_train)

    optimizer.zero_grad()

    loss_calculated.backward()

    optimizer.step()

    if epoch % 20: 
        model2.eval()

        with torch.inference_mode():
            prediction = model2(X_test)
            lossFn(prediction, Y_test)
            print(loss_calculated.item())

    

    
```
